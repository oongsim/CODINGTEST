```python
# 책풀이

n,m,k = map(int,input().split())
arr = [] # 상어 맵
for _ in range(n):
    arr.append(list(map(int,input().split())))
direction = list(map(int,input().split())) # 상어 번호 순서로 방향 담은 리스트 (인덱스0 : 1번상어 ... )
priority = [[] for _ in range(m)] # 상어 번호 순서로 우선 순위 담을 3차원 배열 (인덱스0 : 1번상어 ... )
for i in range(m):
    for _ in range(4):
        priority[i].append(list(map(int,input().split())))
smell = [[[0,0]] * n for _ in range(n)] # [상어번호, 냄새] 저장 리스트
dx = [-1,1,0,0]
dy = [0,0,-1,1]

def smell_update(): # 냄새 업데이트
    for x in range(n):
        for y in range(n):
            if smell[x][y][1] > 0: # 냄새가 있으면
                smell[x][y][1] -= 1 # 1 빼기
            if arr[x][y] != 0: # 상어가 있으면
                smell[x][y] = [arr[x][y], k] # [상어번호, k] 저장
                
def shark_move(): # 상어 이동
    new_arr = [[0]*n for _ in range(n)] # 상어 이동한 뒤 맵 
    for x in range(n):
        for y in range(n):
            if arr[x][y] != 0 : # 상어가 있으면
                d = direction[arr[x][y]-1] # 현재 상어 방향
                found = False # 냄새 없는 곳에서 상어 찾으면 빠져나갈 수 있게 설정해놓은 변수
                
                # 일단 냄새가 존재하지 않는 곳이 있는지 확인
                for i in range(4):
                    nx = x + dx[priority[arr[x][y]-1][d-1][i]-1]
                    ny = y + dy[priority[arr[x][y]-1][d-1][i]-1]
                    if 0 <= nx and nx < n and 0 <= ny and ny < n: # 범위 안에 있고
                        if smell[nx][ny][1] == 0: # 냄새가 존재하지 않는다면 
                            # 해당 상어 방향 업데이트 !! 
                            direction[arr[x][y]-1] = priority[arr[x][y]-1][d-1][i]
                            # 상어 이동시키기
                            if new_arr[nx][ny] == 0: # 옮길 곳에 다른 상어가 존재하지 않는다면
                                new_arr[nx][ny] = arr[x][y]
                            else: # 다른 상어가 존재한다면
                                new_arr[nx][ny] = min(new_arr[nx][ny], arr[x][y]) # 더 작은 상어가 차지
                            found = True
                            break
                if found: # 찾았다면
                    continue # 반복문 빠져나와, 다음 상어 이동시키기
                
                # 아직 못찾앗다면, 주변 냄새가 모두 존재한다는 뜻. 자신의 냄새가 있는 곳으로 이동
                for i in range(4):
                    nx = x + dx[priority[arr[x][y]-1][d-1][i]-1]
                    ny = y + dy[priority[arr[x][y]-1][d-1][i]-1]
                    if 0 <= nx and nx < n and 0 <= ny and ny < n: # 범위 안에 있고
                        if smell[nx][ny][0] == arr[x][y]: # 자신의 냄새가 있다면
                            # 해당 상어 방향 업데이트 !! 
                            direction[arr[x][y]-1] = priority[arr[x][y]-1][d-1][i]
                            # 상어 이동시키기
                            new_arr[nx][ny] = arr[x][y]
                            break                            
    return new_arr

time = 0
while True:
    smell_update()
    new_arr = shark_move()
    arr = new_arr # 상어 맵 업데이트
    time += 1
    
    # 1번 상어가 존재하는지 확인 
    check = True
    for x in range(n):
        for y in range(n):
            if arr[x][y] > 1: # 1번상어 말고 존재한다면
                check = False
    if check: # 1번 상어만 존재하면
        print(time)
        break
    if time > 1000: # 1000초가 넘어도 다른 상어가 존재하면
        print(-1)
        break
```
