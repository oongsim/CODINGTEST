4 3 2 1<br>
0 0 0 0<br>
0 0 9 0<br>
1 2 3 4<br><br>
출력예시3<br>
14<br>
<br><br>

```python
# 책 풀이

from collections import deque

n = int(input())
fish = []
for _ in range(n):
    fish.append(list(map(int,input().split()))) # 물고기 유무와 크기 저장하는 배열
    
now_size = 2 # 아기 상어 현재 크기
now_x, now_y = 0,0

for i in range(n):
    for j in range(n):
        if fish[i][j] == 9: # 아기상어 찾으면
            now_x, now_y = i, j
            fish[i][j] = 0 # 그 위치는 0으로 초기화
                   
dx = [-1,1,0,0]
dy = [0,0,-1,1]
        
# 현재 위치에서 최단 거리 구하기
def bfs():
    dist = [[-1]*n for _ in range(n)] # 최단거리 -1로 초기화
    dist[now_x][now_y] = 0 # 현재 위치 
    q = deque([(now_x,now_y)]) # 큐에 넣기  
    
    while q:
        x,y = q.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx and nx < n and 0 <= ny and ny < n:
                if dist[nx][ny] == -1 and fish[nx][ny] <= now_size: # 처음 방문한 곳이고, 지나갈 수 있다면
                    dist[nx][ny] = dist[x][y] + 1
                    q.append((nx,ny))                  
    return dist # 현재 위치에서 최단 거리 배열

INF = 1e9

# 최단 거리 테이블 주어졌을 때, 먹을 수 있는 물고기 찾기
def find(dist):
    x,y = 0,0
    min_dist = INF
    for i in range(n):
        for j in range(n):
            if dist[i][j] != -1 and 1 <= fish[i][j] and fish[i][j] < now_size: # 도달할 수 있는 거리고, 먹을 수 있는 크기이면
                if dist[i][j] < min_dist: # 가장 가까운 거리에 있는 물고기 하나만
                    x,y,min_dist = i,j,dist[i][j]
                    
    if min_dist == INF: # 해당 조건이 없으면
        return None
    else:
        return x,y,min_dist
    
result = 0
ate = 0
    
while True:
    value = find(bfs())
    if value == None:
        print(result)
        break
    else:
        now_x, now_y = value[0], value[1] # 현재 위치 갱신
        result += value[2] # 이동 거리 변경
        ate += 1
        fish[now_x][now_y] = 0 # 이제 해당 위치에 물고기 없음을 표시 !! 
        if ate == now_size: # 자신의 크기와 같은 수의 물고기 먹으면
            now_size += 1
            ate = 0
```
