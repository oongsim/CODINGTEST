5 8 2 7 13 6 9 2<br><br>
출력예시3<br>
76<br>

```python
# 책 풀이

import copy

arr = [[None]*4 for _ in range(4)] # [번호, 방향] 넣을 2차원 배열

# 방향 , 0부터 7까지
dx = [-1,-1,0,1,1,1,0,-1]
dy = [0,-1,-1,-1,0,1,1,1]

for i in range(4):
    data = list(map(int,input().split()))
    for j in range(4):
        arr[i][j] = [data[2*j], data[2*j+1]-1] # 방향을 인덱스에 맞춰서 설정

def turn_left(direction):
    return (direction+1)%8

def find_fish(arr, num): # 해당 숫자의 위치 인덱스 반환 함수
    for i in range(4):
        for j in range(4):
            if arr[i][j][0] == num:
                return (i,j)
    return None

def fish_move(arr, now_x, now_y):
    for i in range(1,17): # 작은 물고기부터 이동
        position = find_fish(arr,i)
        if position != None:
            x,y = position[0],position[1]
            direction = arr[x][y][1]
            for _ in range(7): # 8번까지 방향 바꾸기 시도 가능
                nx = x + dx[direction]
                ny = y + dy[direction]
                if 0 <= nx and nx < 4 and 0 <= ny and ny < 4: # 해당 범위 안에서
                    if not (nx == now_x and ny == now_y): # 상어가 없다면
                        # 이동가능
                        arr[x][y][1] = direction # 방향이 바뀌었을 경우, 교환 전에 방향 업데이트
                        arr[x][y], arr[nx][ny] = arr[nx][ny], arr[x][y] # 교환
                        break 
                direction = turn_left(direction) # 안되면 방향 바꾸기

def shark_eat(arr,now_x,now_y): # 상어가 먹을 수 있는 물고기 위치 인덱스 반환 함수
    eat_possible = []
    direction = arr[now_x][now_y][1]
    for _ in range(3): # 최대 3개 까지 먹을 수 있는 후보 가능
        now_x += dx[direction]
        now_y += dy[direction]
        if 0 <= now_x and now_x < 4 and 0 <= now_y and now_y < 4: # 해당 범위 안에서
            if arr[now_x][now_y][0] != -1: # 물고기가 존재한다면
                eat_possible.append((now_x, now_y)) # 리스트에 넣기
    return eat_possible

result = 0 # 최종 값 

def dfs(arr, now_x, now_y,total):
    global result
    arr = copy.deepcopy(arr) # 중요 !!!!!!! 리스트 통째로 복사, 만약 최댓값이 아니라면 다시 돌아가야 하므로 !! 
    
    total += arr[now_x][now_y][0] # 상어가 물고기 먹음
    arr[now_x][now_y][0] = -1 # 물고기가 없는 칸은 -1
    
    fish_move(arr,now_x,now_y) # 전체 물고기 움직이고
    
    eat_possible = shark_eat(arr,now_x,now_y) # 상어가 먹을 수 있는 물고기들
    
    if len(eat_possible) == 0: # 종료 조건 !!!!!!!!!!!
        result = max(result, total) # 최대값 저장
        return
    
    for now_x,now_y in eat_possible: # 하나씩
        dfs(arr,now_x,now_y, total)
        
dfs(arr,0,0,0) # 시작
print(result)
```
