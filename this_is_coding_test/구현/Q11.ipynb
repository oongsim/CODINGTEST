```python
# 책 풀이 , 맵 같은 문제는 nx,ny 이용 하자
n = int(input())
maplst = [[0]*(n+1) for _ in range(n+1)]
k = int(input())
for _ in range(k):
    i,j = map(int,input().split())
    maplst[i][j] = 1 # 사과가 있는 곳을 1로 표시
l = int(input())
dturn = []
for _ in range(l):
    x,c = input().split()
    dturn.append((int(x),c)) # 튜플 형태로 방향 변환 정보 저장

# 동, 남, 서, 북 방향
dx = [0, 1, 0, -1]
dy = [1, 0, -1, 0]

# 뱡향 변환 함수
def turn(d, c):
    if c == 'D':
        d = (d+1) % 4
    else:
        d = (d-1) % 4
    return d # 변환된 방향 리턴

# 시뮬레이션
x,y = 1,1 # 처음 뱀 위치
maplst[x][y] = 2 # 뱀이 있는 곳을 2로 표시
q = [(x,y)] # 뱀이 있는 곳을 q라는 리스트에 튜플형태로 표시, 꼬리 제거 하기 위함
time = 0
d = 0 # 현재 방향 
index = 0

while(True):
    nx = x + dx[d] # 다음 가는 칸
    ny = y + dy[d]
    
    if nx >= 1 and nx <= n and ny >= 1 and ny <= n and maplst[nx][ny] != 2: # 범위 밖이 아니고 뱀의 몸이 없다면
        if maplst[nx][ny] == 0 : # 사과가 없다면
            maplst[nx][ny] = 2
            q.append((nx,ny))
            qx,qy = q.pop(0) # q에서 맨 처음 넣었던 꼬리 부분 꺼냄
            maplst[qx][qy] = 0
        else: # 사과가 있다면 
            maplst[nx][ny] = 2
            q.append((nx,ny))
            # 꼬리 부분 제거 하지 않음            
    else: # 범위 밖이거나 뱀의 몸이 있다면
        time += 1 
        print(time)
        break # 게임 끝
    
    time += 1
    x,y = nx,ny
    if index < l and time == dturn[index][0]:
        d = turn(d,dturn[index][1]) # 방향 변환
        index += 1 
        
```
