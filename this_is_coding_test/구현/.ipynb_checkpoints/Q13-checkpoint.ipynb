```python
from itertools import combinations

n,m=map(int,input().split())
maplst = []
for _ in range(n):
    maplst.append(list(map(int,input().split())))

# jips, chickens 리스트에 좌표 입력 
jips = []
chickens = []
for i in range(n):
    for j in range(n):
        if maplst[i][j] == 1:
            jips.append((i,j))
        elif maplst[i][j] == 2:
            chickens.append((i,j))
            
# 치킨집에서 k개 뽑는 조합 만들기
chickens_combi = list(combinations(chickens,m))

distance = 100 # 최소거리 임의 설정
for chicks in chickens_combi: # chicks는 2차원 튜플 형태임
    result = []
    for jip in jips: # jip, chicken 1차원 튜플 형태
        d = 100 # 최소거리 임의 설정
        for chicken in chicks:
            hd = abs(jip[0]-chicken[0]) + abs(jip[1]-chicken[1]) # 현재 거리
            d = min(hd, d) # 치킨거리 구하기
        result.append(d) # 각 집의 치킨거리를 result 리스트에 넣기
    dist = sum(result) # 치킨거리 값
    distance = min(dist, distance) # 조합 중에서 치킨거리의 최솟값 구하기
print(distance)
```

```python
# 책 풀이. 입력 받을 때 아예 리스트에 저장

from itertools import combinations

n,m = map(int,input().split())
chicken, house = [], []

for r in range(n):
    data = list(map(int,input().split()))
    for c in range(n):
        if data[c] == 1:
            house.append((r,c)) # 일반 집
        elif data[c] == 2:
            chicken.append((r,c)) # 치킨집
            
# 모든 치킨집 중에서 m개의 치킨집을 뽑는 조합 계산
candidates = list(combinations(chicken, m))

# 치킨 거리의 합을 계산하는 함수
def get_sum(candidate):
    result = 0
    # 모든 집에 대하여
    for hx, hy in house:
        # 가장 가까운 치킨집 찾기
        temp = 1e9
        for cx, cy in candidate:
            temp = min(temp, abs(hx-cx)+abs(hy-cy))
        # 가장 가까운 치킨집까지의 거리를 더하기
        result += temp
    # 치킨 거리의 합 반환
    return result

# 치킨 거리의 합의 최소를 찾아 출력
result = 1e9
for candidate in candidates:
    result = min(result, get_sum(candidate))
    
print(result)
```
