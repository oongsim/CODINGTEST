```python
import time
from itertools import combinations

n,m = map(int,input().split())
maplst = []
for _ in range(n):
    maplst.append(list(map(int,input().split())))
    
    
start_time = time.time() # 측정 시작
    
    
bucks = [] # 벽이 될 수 있는 리스트
for i in range(n):
    for j in range(m):
        if maplst[i][j] == 0: # 빈칸이면
            bucks.append((i,j))
            
def dfs(i,j): # i,j는 인덱스
    # 주어진 범위를 벗어난 경우 즉시 종료
    if i < 0 or i >= n or j < 0 or j >= m:
        return
    # 빈칸일 때 바이러스 감염
    if maplst[i][j] == 0:
        maplst[i][j] = 3 # 감염시킨 곳은 3으로 바꿔줌
        # 주변도 dfs 
        dfs(i-1,j)
        dfs(i+1,j)
        dfs(i,j-1)
        dfs(i,j+1)
        return 
    return 
    
maxsize = 0
for buck in list(combinations(bucks,3)): # 벽 3개의 조합 중 하나씩 ((i,j),(i,j),(i,j)) 형태
    cnt = 0
    for b in buck: # (i,j) 형태
        x,y = b
        maplst[x][y] = 1 # 벽으로 만들 부분 1로 바꾸기
    
    # dfs 수행
    for i in range(n):
        for j in range(m):
            if maplst[i][j] == 2: # 바이러스 일때 dfs 수행
                dfs(i-1,j)
                dfs(i+1,j)
                dfs(i,j-1)
                dfs(i,j+1)
                
    # 0의 개수 세기
    for i in range(n):
        for j in range(m):
            if maplst[i][j] == 0: # 0이면
                cnt += 1
            if maplst[i][j] == 3: # 감염시킨 곳 다시 0으로 초기화 
                maplst[i][j] = 0
        
    maxsize = max(maxsize, cnt) # 0으로 된 부분 최대 값
        
    for b in buck:
        x,y = b
        maplst[x][y] = 0 # 벽으로 만들어줬던 부분 다시 0 으로 
        
print(maxsize)


end_time = time.time() # 측정 종료
print("time :", end_time - start_time) # 수행 시간 출력
```

```python
import time

# 책 풀이, 맵 리스트를 2개 만들어서 진행, 울타리 설치도 dfs 사용, 근데 내가 짠 코드가 시간이 훨씬 적게 걸리는 듯.. 

n,m = map(int,input().split())
data = [] # 초기 맵 리스트
temp = [[0]*m for _ in range(n)] # 벽을 설치한 뒤의 맵 리스트

for _ in range(n):
    data.append(list(map(int,input().split())))
    
    
start_time = time.time() # 측정 시작

    
# 4가지 이동 방향에 대한 리스트
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

result = 0

# 깊이 우선 탐색(DFS)을 이용해 각 바이러스가 사방으로 퍼지도록 하기
def virus(x,y):
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        # 상, 하, 좌, 우 중에서 바이러스가 퍼질 수 있는 경우
        if nx >= 0 and nx < n and ny >= 0 and ny < m:
            if temp[nx][ny] == 0:
                # 해당위치에 바이러스  배치하고, 다시 재귀적으로 수행
                temp[nx][ny] = 2
                virus(nx,ny)
                
# 현재 맵에서 안전 영역의 크기 계산하는 메서드
def get_score():
    score = 0
    for i in range(n):
        for j in range(m):
            if temp[i][j] == 0:
                score += 1
    return score

# 깊이 우선 탐색(DFS)을 이용해 울타리를 설치하면서, 매번 안전 영역의 크기 계산
def dfs(count):
    global result
    
    # 울타리가 3개 설치된 경우
    if count == 3:
        for i in range(n):
            for j in range(m):
                temp[i][j] = data[i][j] # 현재 맵을 임시 맵으로 지정
        # 각 바이러스 위치에서 전파 진행
        for i in range(n):
            for j in range(m):
                if temp[i][j] == 2:
                    virus(i,j)
        # 안전 영역의 최대값 계산
        result= max(result, get_score())
        return 
    
    # 빈 공간에 울타리 설치
    for i in range(n):
        for j in range(m):
            if data[i][j] == 0:
                data[i][j] = 1
                count += 1
                dfs(count)
                data[i][j] = 0 # 다시 원상복귀
                count -= 1
                
dfs(0)
print(result)

end_time = time.time() # 측정 종료
print("time :", end_time - start_time) # 수행 시간 출력
```
