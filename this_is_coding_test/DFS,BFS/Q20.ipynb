# 감시 피하기
---
입력 예시<br>
5<br>
X S X X T<br>
T X S X X<br>
X X X X X<br>
X T X X X<br>
X X T X X<br>
<br>
출력 예시<br>
YES<br>
<br><br>
입력 예시<br>
4<br>
S S S T<br>
X X X X<br>
X X X X<br>
T T T X<br>
<br>
출력 예시<br>
NO<br>
<br><br>

```python
# 책 풀이, 조합 이용, DFS 이용해서 풀던거 날라감.. ㅠ 근데 이 조합이 더 쉬운 방법 같음 그냥 풀던대로 조합으로 풀기.. 
from itertools import combinations

n = int(input())
maplst = []
teachers = [] # 선생님 위치 정보
spaces = [] # 빈 공간 위치 정보

for i in range(n):
    maplst.append(list(input().split()))
    for j in range(n):
        if maplst[i][j] == 'T': # 선생님이 존재하는 위치 저장
            teachers.append((i,j))
        if maplst[i][j] == 'X': # 빈 공간 위치 저장
            spaces.append((i,j))
            
# 특정 방향으로 감시를 진행 (학생 발견: True, 학생 미발견: False)
def watch(x, y, direction):
    # 왼쪽 방향으로 감시
    if direction == 0:
        while y >= 0:
            if maplst[x][y] == 'S': # 학생이 있는 경우
                return True
            if maplst[x][y] == 'O': # 장애물이 있는 경우
                return False
            y -= 1
    # 오른쪽 방향으로 감시
    if direction == 1:
        while y < n:
            if maplst[x][y] == 'S': # 학생이 있는 경우
                return True
            if maplst[x][y] == 'O': # 장애물이 있는 경우
                return False
            y += 1
    # 위쪽 방향으로 감시
    if direction == 2:
        while x >= 0:
            if maplst[x][y] == 'S': # 학생이 있는 경우
                return True
            if maplst[x][y] == 'O': # 장애물이 있는 경우
                return False
            x -= 1  
    # 아래쪽 방향으로 감시
    if direction == 3:
        while x < n:
            if maplst[x][y] == 'S': # 학생이 있는 경우
                return True
            if maplst[x][y] == 'O': # 장애물이 있는 경우
                return False
            x += 1
    return False

# 장애물 설치 이후에, 한 명이라도 학생이 감지 되는지 검사
def process():
    # 모든 선생님의 위치를 하나씩 확인
    for x, y in teachers:
        # 4가지 방향으로 학생을 감지할 수 있는지 확인
        for i in range(4):
            if watch(x,y,i): # 학생 감지가 되었으면
                return True
        return False # 4방향 모두 학생 감지가 되지 않으면

result = False # 결과

# 빈 공간에서 3개를 뽑는 모든 조합을 확인
for data in combinations(spaces, 3):
    # 장애물 설치해보기
    for x, y in data:
        maplst[x][y] = 'O'
        
    # 학생이 모두 감지되지 않은 경우
    if not process():
        # 원하는 경우를 발견한 것임
        result = True
        break
    # 설치된 장애물을 다시 없애기
    for x, y in data:
        maplst[x][y] = 'X'
        
if result == True:
    print('YES')
else:
    print('NO')
```
